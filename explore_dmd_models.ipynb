This notebook plots the spatial modes and dynamics of a DMD model fitted to a 10 year slice of atmospheric temperature at 850 hPa from the ERA5 dataset.

```python
import pickle
import cartopy.crs as ccrs
from matplotlib import pyplot as plt
import numpy as np
```

Open the DMD model saved to disk.

```python
with open("models/deterministic/dmd_10.pkl", "rb") as f:
    dmd = pickle.load(f)
```

```python
modes = dmd.modes.unstack()
```

Sort the modes by amplitude.

```python
mode_order = np.argsort(-np.abs(dmd.amplitudes))
```

Plot the first DMD mode.

```python
plt.rcParams.update({'font.size': 14})
fig, ax = plt.subplots(figsize=(15, 5), subplot_kw={'projection': ccrs.PlateCarree()})

# plot the real part of the DMD mode with highest amplitude
modes[mode_order[0], :, :].real.plot.contourf(
    ax=ax, transform=ccrs.PlateCarree(), cmap='coolwarm',
    vmin=-0.002, vmax=0.002,
    cbar_kwargs={"label": ""},
)
ax.coastlines(color='black')
ax.gridlines(draw_labels=True)

plt.title("First DMD mode of T850")
plt.show()
```

```python
# plot the real part of the temporal evolution of the first DMD mode
plt.rcParams.update({'font.size': 14})
dmd.dynamics[mode_order[0], :].real.plot(figsize=(12, 4))
plt.title("Dynamics of first DMD mode of T850")
plt.ylabel("")
plt.show()
```

Calculate the oscillation period.

```python
hours_per_day = 24
oscillation_period = 2 * np.pi / abs(dmd.eigs[mode_order[0]].imag) / hours_per_day
print(f"Oscillation period of 1st DMD mode: {oscillation_period:.2f} days")
```

Now plot the third DMD mode.

```python
plt.rcParams.update({'font.size': 14})
fig, ax = plt.subplots(figsize=(15, 5), subplot_kw={'projection': ccrs.PlateCarree()})

# plot the real part of the DMD mode with highest amplitude
modes[mode_order[2], :, :].real.plot.contourf(
    ax=ax, transform=ccrs.PlateCarree(), cmap='coolwarm',
    vmin=-0.002, vmax=0.002,
    cbar_kwargs={"label": ""},
)
ax.coastlines(color='black')
ax.gridlines(draw_labels=True)

plt.title("Third DMD mode of T850")
plt.show()
```

```python
# plot the real part of the temporal evolution of the first DMD mode
plt.rcParams.update({'font.size': 14})
dmd.dynamics[mode_order[2], :].real.plot(figsize=(12, 4))
plt.title("Dynamics of third DMD mode")
plt.ylabel("")
plt.show()
```

```python
hours_per_day = 24
oscillation_period = 2 * np.pi / abs(dmd.eigs[mode_order[2]].imag) / hours_per_day
print(f"Oscillation period of 3rd DMD mode: {oscillation_period:.2f} days")
```
